```python
nyc_check_in = gpd.read_file(r'shp/nyc_checkin_sticc3.shp',engine='pyogrio')
nyc_check_in.head(1)
```

```python
nyc_check_in.groupby("venueCateg").count().sort_values("venueId").tail(1)
```

```python
venueCateg_list = ["Office", "Home (private)"]
venueId_list = pd.DataFrame(nyc_check_in.venueId.unique()).sample(frac=0.3).values.squeeze()
nyc_check_hdcbc = nyc_check_in[(nyc_check_in.venueCateg.isin(venueCateg_list))&(nyc_check_in.venueId.isin(venueId_list))]
print(nyc_check_hdcbc.shape)
nyc_check_hdcbc.head(1)
```

```python
def return_week(x):
    if x == "Mon":
        return 1
    elif x == "Tue":
        return 2
    elif x == "Wed":
        return 3
    elif x == "Thu":
        return 4
    elif x == "Fri":
        return 5
    elif x == "Sat":
        return 6
    elif x == "Sun":
        return 7
    
def return_category(x):
    if x == "Gym":
        return 1
    elif x == "Coffee Shop":
        return 2
    elif x == "Office":
        return 3
    elif x == "Home (private)":
        return 4
    elif x == "Subway":
        return 5

nyc_check_hdcbc["week_attr"] = nyc_check_hdcbc["week"].apply(lambda x: return_week(x))
nyc_check_hdcbc["category"] = nyc_check_hdcbc["venueCateg"].apply(lambda x: return_category(x))
nyc_check_hdcbc = nyc_check_hdcbc.reset_index().drop("index", axis=1)
```

```python
nyc_check_hdcbc[["latitude","longitude","week_attr", "hour","category"]].to_csv(
    r'dataset/nyc_checkin.csv', 
    header=True,  
    index=False,  
    sep=','
)
```

```python

def noise_generate(data):

    original_data=data.values
    x_min, x_max = original_data[:, 0].min(), original_data[:, 0].max()
    y_min, y_max = original_data[:, 1].min(), original_data[:, 1].max()


    length=int(original_data.shape[0]/10)
    noise_x = np.random.uniform(x_min, x_max, length)
    noise_y = np.random.uniform(y_min, y_max, length)
    noise_label = np.full(length, -1)

    #print(original_data)
    nose=np.column_stack((noise_x, noise_y, noise_label))

    noisy_data = np.vstack((original_data,nose))
    pdf=pandas.DataFrame(noisy_data)

    return pdf
```

```python
import pickle
import pandas

folder_path=r'dataset/'

file_list = os.listdir(folder_path)


data={}
for file_name in file_list:
    file_path = os.path.join(folder_path, file_name)
    if file_name.startswith('.'):
        continue
    file_name=file_name.replace('.csv','')

    print(file_path)

    if os.path.isfile(file_path):

        try:
            d=pandas.read_csv(file_path)
            
           
            # columns_to_keep = [col for col in d.columns if col not in ["week_attr", "hour"]]

           
            # d = d[columns_to_keep]
            
            
            #print(d.values[:,0].min())

            #d=noise_generate(d)


            data[file_name]=d
            print(d.head())
        except ZeroDivisionError as e:

            print("Division by zero:", e)
        except ValueError as e:

            print("Value error:", e)
        except Exception as e:

            print("An error occurred:", e)

data_list={}
for key in data.keys():

    data_list[key] = data[key]




print(len(data_list))

with open(r'cache/data_csv.pkl','wb') as file:
    pickle.dump(data_list,file)

```
